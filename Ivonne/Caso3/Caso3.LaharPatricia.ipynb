## Introducción

#### Sismología ambiental 
#### Clásicamente la sismología era empleada para localizar eventos tectónicos, pero estos eventos sólo representaban el 1% de la señal registrada _(Burtin et al 2016)_ la información restante correspondía al llamado "ruido ambiental" que, con el tiempo fue objeto de interés ya que se observó que su análisis permite la caracterización de otro tipo de eventos geológicos superficiales, de esta manera surge la sismología ambiental.
#### La sismología ambiental consiste en estudiar las vibraciones mecánicas que se originan o han sido afectadas por causas externas, es decir, causas externas a la Tierra Sólida _(Larose et al. 2015)_. Dichas causas externas tienen lugar en procesos superficiales relacionados a la naturaleza (movimientos en masa, flujos, eventos hidrometeorológicos y criogénicos) y a la acción humana (explotación de pozos, pruebas nucleares, construcciones, etc.) , el punto de interés en el estudio de la simología ambiental radica en sistemas de monitoreo de procesos superficiales que representan una fuente de peligro para las comunidades y monitoreo sitios de explotación de recursos. Escencialmente, la diferencia entre sismología clásica y sismología ambiental es el tipo de fuente que produce la señal, pues el procesamiento y análisis de datos suele ser el mismo.

#### Lahares
#### Dentro de la lista de procesos superficiales que pueden caracterizarse através de señales sísmicas se encuentran los flujos de escombros o lahares, una definición formal de lahar es “un flujo altamente concentrado de una mezcla de rocas, lodo y agua, proveniente de un volcán” _(Schmidt 1934 y Bemmelen 1949)_. Cuantitativamente, a un flujo se le puede llamar lahar si tiene una concentración en volumen de sedimento 47-77% _(Costa, 1987)_, o bien un contenido agua/sedimento proporcional _(Vallance, 2005)_, cabe resaltar que los lahares no mantienen un comportamiento meramente de flujo de detritos a lo largo de todo el evento, debido a los cambios en el volumen agua/sedimento, por tasas diferenciales de sedimentación/depositación, interacción con la morfología del canal y asimilación de material a su paso. Un lahar puede ser descrito (generalmente) como un evento de tres fases _(Vallance, 2005)_. :

+ flujo de corriente: contiene < 20% de volumen de sedimento, la fracción presente es de grano fino que se mueve en suspensión, el agua es la fase continua del flujo y las partículas no interactúan entre sí, se comporta como un fluido Newtoniano, turbulento _(Costa, 1978)_.
+ flujo hiperconcentrado (transicional): es la fase intermedia entre un flujo de corriente y uno de detritos, contiene mayor porcentaje de sedimento fino que el flujo anterior y se incorporan fracciones gruesas (%Vol 20-60%), existe una separación vertical del material y tiene un comportamiento no-Newtoniano, entre turbulento y laminar _(Costa, 1987)_.
+ flujo de detritos: el porcentaje de volumen de detritos es 70-90%, es un flujo cohesivo que puede soportar grandes bloques en suspensión, se considera un fluido tipo Bingham (viscoplástico) _(Costa, 1987)_, tipo laminar, éste tipo de flujo favorece la depositación a diferencia de los dos anteriores ya que se consideran mayormente erosivos _(Vallance, 2005)_.


#### Los lahares tienen morfologías características facilitando su representación en tres partes (Figura 1, A) consiste en: un frente, un cuerpo y una cola. El frente se caracteriza por ser la parte frontal del flujo que lleva la mayor cantidad de sedimento y bloques de mayor tamaño que se transportan generalmente por empuje y asimilan material a su paso (proceso de bulking) recirculando fragmentos, el cuerpo es una zona donde existe reciclo y transporte horizontal del material dentro del flujo, hay una relación agua/sedimento más alta y el transporte lo domina el soporte del agua (flujo cohesivo), finalmente la cola es la parte más diluida del flujo y se caracteriza por tener concentración en volumen de sedimento pequeña, < 20%.

#### Mecanismos de formación y disparo
#### La formación de un lahar requiere esencialmente una gran cantidad de agua, material no consolidado, pendientes altas y un mecanismo de disparo _(Costa, 1987)_. El origen del agua puede darse por derretimiento rápido de hielo, nieve, agua subglacial, ruptura de lago cráter, otro tipo de lago, lluvias torrenciales y aporte hidrotermal _(Costa, 1987)_. Los mecanismos de formación de lahares pueden ser syn-eruptivos (durante una erupción, asociados a PDS (Pyroclastic Density Currents)), post-eruptivos (después de una erupción) e intraeruptivos (sin relación directa con erupciones), los disparadores son generalmente lluvias intensas, sismos volcánicos, explosiones, e intrusiones.  Un evento de lahar también puede ser inducido por la ocurrencia de otros eventos superficiales como avalanchas de escombros que durante su transporte interactúan con fuentes de agua, se saturan y transforman en flujos de detritos (volcánicos y exóticos).
#### Para el volcán de Colima, la fuente de agua es el aporte de lluvias estacionales que a su vez funciona como un mecanismo disparador, se han observado eventos post-eruptivos _(Capra et al. 2016)_ e intraeruptivos, donde el origen del material ha sido depósitos de PDC’s (generación de lahares calientes), depósitos de erupciones pasadas, depósitos de lahares anteriores y material aluvial remanente en los canales.

#### Importancia
#### Debido a que la zona de estudio comprende un volcán activo y pertenece a una región de temporada de secas y lluvias muy definida, los factores de formación y disparo están presentes y son comunes, históricamente, se han observado distintos eventos que han representado un riesgo a las poblaciones aledañas, instalaciones eléctricas y vías de comunicación _(Capra et al. en prensa)_. En octubre del año 1958 un flujo de escombros disparado por fuertes lluvias estacionales viajó aprox 14 km desde la ladera noreste del Volcán de Colima hasta la parte sureste del estado de Jalisco y sepultó parte el poblado de Atenquique y detuvo sus actividades durante varios meses, en 2011, 2013 y 2015 se produjeron eventos de lahares (también asociados a lluvias estacionales), es por esto que un sistema de monitoreo representa una forma de mitigación y prevención de riesgo. Cabe mencionar que la implementación de sistemas de monitoreo tienen el objetivo principal de generar un sistema de alerta temprana a través del análisis de datos y caracterización de los eventos para obtener una predicción de los mismos. 

#### Sistemas de monitoreo
#### La sismicidad juega un papel fundamental en los sistemas de alerta temprana _(Moreno, 2015)_, pues es un método que permite la  identificación de procesos através de características espectrales y automatización de algoritmos de caracterización para la toma de desiciones en tiempo corto. Debido al peligro latente que representan los eventos generados y asociados al volcán de Colima se han implementado sistemas de monitoreo desde el año 2007 _(Davila et al., 2007; Zobin et al., 2009; Capra et al., 2010)_, actualmente están establecidas dos estaciones de monitoreo en la barranca La Lumbre y la barranca Montegrande. Ambas estaciones consisten en una torre de 12m de altura con una antena direccional que transmite datos en tiempo real a las instalaciones de la RESCO, una videocámara que toma imágenes cada 2-4 segundos con una resolución de 704 x 480 pixeles, un pluviómetro (HOBO RG3) con un sensor de humedad y un geófono de 10Hz (Sercel) _(Vázquez et al., 2016a; Coviello et al., en prensa)_. Además, en la estación de Montegrande se cuenta con un sensor de infrasonido y otro de banda ancha.  Para el presente análisis se utilizaron las señales obtenidas del geófono Sercel de 10 Hz  de la estación en la barranca La Lumbre y datos de lluvia obtenidos de una estación meteorológica del estado de Jalisco ubicada en el volcán nevado de Colima, las lecturas de descarga de lluvia (en milímetros) cada diez minutos, a diferencia del pluviómetro de las estaciones del volcán de Colima que toman registro cada minuto.

#### Señales de lahares
#### La vibraciones en el suelo causadas por flujos de escombros se caracterizan por tener transiciones abruptas, discontinuidades de salto, amortiguación y rasgos no estacionarios _(Huang et al 2007)_, los lahares producen señales sísmicas de larga duración que se parecen a los tremores volcánicos, excepto porque la señal es más fuerte cerca del canal de flujo _(McNutt, 2002)_, pueden diferenciarse de otros eventos superficiales (PDC´s) por su duración (de pocos minutos a horas), contenido de altas frecuencias (10-60Hz, hasta 100 Hz en algunos casos) y velocidades menores _(Thompson, 2015)_, según _Zobin, 2012_, el rango de frecuencias respectivo a señales de lahares es 10-25Hz, lo cual podría discriminar eventos con comportamientos distintos a los observados. En la Figura 2 se puede apreciar un evento de flujo de detritos que ha sido identificado por sus características espectrales (duración, banda de frecuencias y distribución de la energía en función de las mismas).
#### A partir de una serie de experimentos realizados por _Huang et al 2007_ y el análisis de un evento de flujo de detritos ocurrido en Ai-Yu-Zi, Taiwan, se observó que el rango de frecuencias a lo largo de un evento de éste tipo es mayor y variable (a diferencia de los anteriormente mencionados) conforme a la evolución y composición del flujo, se determinó que para la porción frontal del flujo caracterizada por tener la mayor cantidad de fragmentos sólidos y de mayor tamaño; el rango de frecuencias es menor de 50 Hz, después del paso del frente, la banda de frecuencias es mayor: de 50-100 Hz y en el momento que el flujo pasa por el sensor el rango de frecuencias es muy amplio 10-250 Hz. Así mismo, se reporta que, para fragmentos de roca con grandes medidas y pesos, el pico de frecuencias migra hacia valores más bajos en contraste con fragmentos más pequeños y ligeros que dan lugar a picos en frecuencia de valores más altos, lo que es muy útil para determinar la composición y dinámica del flujo, además se sugiere que el mayor aporte de vibraciones tiene origen en la colisión y fricción que ocurre entre las rocas y el lecho del canal.
#### La clasificación del registro sísmico de un lahar no es trivial, pues intervienen factores como la atenuación, tomando en cuenta que es un proceso superficial en el que interactúa un fluido con alta concentración de partículas de tamaños y formas aleatorias que pueden producir frecuencias altas que a su vez son rápidamente atenuadas durante su interacción con el sustrato, además de ser una fuente móvil y fluctuante, es importante señalar que un sistema de monitoreo de lahares debe ser un conjunto de métodos que permitan la caracterización del evento y la estación debe colocarse en un sitio clave que permita la mejor recepción de datos. Se puede adaptar una red sismo-volcánica existente para el monitoreo de lahares añadiendo estaciones cercanas al canal de flujo en varios puntos, permitiendo que las señales del lahar se resuelvan más fácilmente y se ubiquen crudamente utilizando los tiempos de las amplitudes máximas en cada estación, y también la distribución de la amplitud a través de la red sísmica _(Thompson, 2015)_.

```julia
# Función promedio móvil
function MA(x,n)
    if size(x,1)==1
        x = x'
    end
    y = zeros(length(x))
    sx = size(x,2)
    tape = NaN*(zeros(convert(Int,floor(n/2)),sx))
    x1 = [tape;x;tape]
    n1 = n-1
    for ii=1:size(y,1)
        sel = x1[ii+(0:n1),:]
        y[ii]=mean(sel[!isnan.(sel)]);
    end
    return y
end
```

```julia
pro = MA(env1,400);
```

## Análisis tiempo-frecuencia, Transformada Rápida de Fourier
### La amplitud sísmica de una serie de tiempo puede ofrecer información relevante sobre una señal sísmica pero algunas características importantes pueden enmascararse por medio de entradas que no están relacionadas con el evento de interés _(Burtin, et al. 2016)_ . La transformación de la señal a un espectro de frecuencias es una herramienta útil para la caracterización de la señal y se logra através de un análisis tiempo-fecuencia, el método más común de dicho análisis es la trasformada de Fourier (TF). La transformada de Fourier es una función que manipula la señal en el dominio del tiempo para pasar al dominio de frecuencia  através de la función (1), para que una TF exista se deben cumplir las siguientes condiciones: que exsista la integral del valor absoluto de la señal _s(t)_, que las discontinuidades en _s(t)_ sean finitas y que _s(t)_ sea de variación acotada _(Nava, 2002)_. La aplicación de una transformada de Fourier divide la señal en una serie de segmentos y aplica la función:
### $ FT[s(t)] = s(t),\exp^{i2\pi ft} = \int_{-\infty}^{\infty} s(t) \exp ^{-2\pi ft} dt $ (1)
### donde _t_ es el tiempo _f_ es la frecuencia y _i_= $\sqrt {-1}$  *(Tary et al. 2014)*. Cabe señalar, que la transformada de Fourier se aplica a señales que se asumen estacionarias , la transformada  de Fourier rápida (_fft_) es la extensión más común para manejar señales de tiempo no estacionarias, se puede expresar como:
### S<sub>f $(\tau, f) = \int_{-\infty}^{\infty} s(t)w(t-\tau) \exp ^{-i2\pi ft} $ $(2)$
### que corresponde a la transformada de Fourier de una señal _s(t)_ reducida por la función _w(t)_ donde $\tau$ es el tiempo de retraso al centro del _taper_, la función $(2)$ es un caso particular de la función $(1)$ donde se usa la ventana _w(t)_ para seleccionar la parte de _s(t)_ que se transformará _(Tary et al. 2014)_. La _fft_ por lo tanto, es simplemente un algoritmo que aprovecha simetrías y periodicidades de las funciones trigonométricas para evaluar una transformada de Fourier discreta _(Nava, 2002)_.

```julia
#Espectros: transformada rápida de fourier
esp1 = abs.(fft(pa1.t,1));
esp2 = abs.(fft(pa2.t,1));
```

```julia
fig = figure(figsize=(12,8))
subplot(2,1,1)
title("Figura 8: Espectro")
plot(f1[1:n1],esp1[1:n1],"c",label = "100sps")
ylabel("Amplitud del espectro")
legend(loc="best")
subplot(2,1,2)
plot(f2[1:n2],esp2[1:n2],"g",label = "250sps")
xlabel("Frecuencia[Hz]")
ylabel("Amplitud del espectro")
legend(loc="best")
```

## Filtro Butterworth y método de Welch
### Los filtros son herramientas matemáticas o instrumentales que permiten la restricción del paso (filtros a priori: antes del registro) de señales que se consideran ruido (señal de no interés) o la eliminación de valores (filtros a posteriori: después del registro) que se encuentran fuera de la señal de interés o afectan a la misma por factores ajenos al evento en cuestión, los más comunes son los filtros pasabajas, pasa altas y pasabandas.
### El filtro digital Butterworth es uno de los más comunes de tipo pasa bajas, se calcula como la relación: 
###                                                   $|{B(w)}| = \frac{1}{1+(\frac{w}{w_p})^{2N}}$
### Donde:
 ### _w_ es la frecuencia registrada
 ### _wp_ es la frecuencia de corte
 ### N es el orden del filtro

### Se ha aplicado la banda 6-50 y 6-125 Hz para los datos de 100sps y 250sps respectivamente apartir del reconocimiento de la respuesta instrumental (Figura 4).

### Método de Welch
### Alternativo al cálculo del espectro de la señal por la _fft_ el método de superposición de Welch es otro estimador que permite el análisis tiempo-frecuencia y la obtención de un espectro de densnsidad de potencias (_PSD Power Spectral Density_), en el que una serie de tiempo se divide en varios segmentos superpuestos (traslape) y se calcula un PSD para cada segmento, se determina un PSD promedio para todos los segmentos, que sirve para reducir la varianza del espetro final _(Burtin et al. 2016)_. Sin embargo, es importante mencionar que el método de Welch disminuye la resolución del espectro en frecuencias bajas (al rededor de 1Hz).

```julia
#Filtro: respuesta del instrumento, apartir de 10 Hz, pasabandas de 6-50/125Hz
resp1 = Bandpass(6,49,fs=100)
desig = Butterworth(4)
fil1 = filt(digitalfilter(resp1,desig),pa1.t);

resp2 = Bandpass(6,124,fs=250)
desig = Butterworth(4)
fil2 = filt(digitalfilter(resp2,desig),pa2.t);
```

```julia
#Cálculo del espectro por el método welch
filt1 = welch_pgram(fil1,fs=100); #Pobar Bartlett
filt2 = welch_pgram(fil2,fs=250);
```

```julia
fig = figure(figsize=(12,8))
subplot(2,1,1)
title("Figura 9: Señal filtrada")
plot(filt1.freq,filt1.power,"c", label="100sps")
ylabel("PSD")
legend(loc="best")
subplot(2,1,2)
plot(filt2.freq,filt2.power,"g", label="250sps")
xlabel("Frecuencia[Hz]")
ylabel("PSD")
legend(loc="best")
```

## Espectrograma
### Otra parte del análisis de tiempo-frecuencia através de los métodos de Fourier es el cálculo de un espectrograma por medio de la transformada de Fourier discreta (_DFT por sus siglas en inglés Discrete Fourier Transform)_ a los espectros variables en el tiempo, también es llamada la transformada de Fourier de corto tiempo (_STFT Short-Time Fourier Transform_) (Tary, et al. 2014), éste gráfico permite la visualización de la distribución de la energía sísmica en tiempo y frecuencia, ayuda a determinar las caracerísticas principales de un evento geomorfológico, dichas características se pueden utilizar apriori en un proceso automatizado para la identificación de fuentes geomorfológicas y para limitar la ocurrencia de falsas alarmas, crucial en el desarrollo de sistemas de alerta _(Burtin et al. 2016)_. 

```julia
#Espectrograma
fig = figure(figsize=(12,6))
subplot(2,1,1)
title("Figura 10 : Espectrograma [dB/Hz] ")
specgram(pa1.t,50,100,pad_to=40,noverlap=49);
ylabel("Frecuencia[Hz]")
colorbar()
subplot(2,1,2)
specgram(pa2.t,125,250,pad_to=40,noverlap=50);
xlabel("Tiempo[s]")
ylabel("Frecuencia[Hz]")
ylim(0,50)
colorbar()
```

## Discusiones
### A partir del análisis en tiempo; visualización de la señal (Figura 5) y su envolvente suavizada (Figura 7); se identificaron tres picos (Figura 7.1) de mayores amplitudes: a1 (0.0026 m/s), a2 (0.0020 m/s) y a3 (0.0075 m/s) (Figura 15, exageración de la señal por 1000), los dos primeros picos corresponden a pasos de dos frentes identificados en las fotografías (Imágenes 4 y 6 de la Figura 14), tanto a1 como a2 se relacionan con una ligera disminución en la curva de descarga de lluvia (Figura 12.1, exageración de la envolvente por 3000), el tercer máximo en amplitud a3 corresponde al punto de mayor velocidad de la señal y mayor descarga de lluvia durante el evento: 0.0075 m/s y 7.2mm respectivamente, no se cuenta con imágenes para correlacionarlo debido a la hora (04:00 24/10/15 GMT, 7pm 23/10/15 tiempo real) pero sugiere que es el paso de un tercer pulso mucho más energético que los anteriores. La Figura 11, correlaciona la descarga de lluvia con la envolvente sin suavizar, es notable que la variación en picos máximos es mayor por lo que para este análisis se tomó en cuenta la envolvente suavizada. Por otra parte, el análisis espectral tiempo-frecuencia permitió definir un rango de frecuencias característico del evento ~10-25 Hz (tanto para la señal muestreada a 100sps como la de 250sps, aunque puede ser más claro en la primera gráfica), que coincide con los rangos reportados por _Zobin, 2012_ y _Thompson, 2015_, el pico más alto entre el rago de frecuencias es ~12Hz (Figuras 8 y 9), en la figura 10, que corresponde al espectrograma es claro que en la banda de frecuencias mencionada se encuentra la concentración de la mayor parte de la energía liberada durante el evento (<-75dB/Hz) que a su vez, corresponde con los máximos identificados en tiempo. 
### De acuerdo con Capra et al, en discusión, la respuesta hidrológica de la barranca a la formación de lahares se da apartir de una acumulación de al rededor del 40% del evento de lluvia, al inicio del evento ya se habían acumulado 255 mm de 440.2mm (>50%) por lo que la respuesta de la barranca a este evento coincide. Cabe mencionar que no se observan diferencias relevantes entre la señal obtenida a partir de un muestreo a 100sps o a 250 sps por lo que para el análisis final de correlación sólo se usó el gráfico de la señal y envolvente suavizada de los datos a 100sps.

## Conclusiones
### Tomando en cuenta las características observadas: duración (7hrs), banda de frecuencias (~10-25/30 Hz), evolución, el análisis de la señal en los dos dominios presentados puede discriminar otro tipo de proceso superficial, además de contar con imágenes que sustenten el comportamiento del lahar. A partir del análisis de la misma señal a diferente frecuencia de muestreo se pudo evidenciar que no es necesaria una tasa de muestreo mayor a 100sps para este evento, pues la recuperación en bandas de frecuencia corresponden a las características del evento y no se registró una evolución destacable en frecuencias mayores a 30 Hz, sin embargo, se han reportado rangos de frecuencias mayores en flujos más complejos _(Huang et al. 2007)_ que podrían tener lugar en la zona de estudio tomando en cuenta las condiciones de formación y disparo, por lo que sí sería necesario implementar una mayor frecuencia de muestreo.
### La identificación de este tipo de eventos a través de su registro sísmico en tiempo real, permite la generación de sistemas de alerta temprana, además, para el caso de estudio conociendo los mecanismos formadores y disparadores se puede generar la automatización de dicha alerta por medio de una predicción del evento en función de la cantidad de descarga y acumulación de lluvia así como evolución en la amplitud de la señal, de esta forma se pueden generar planes de mitigación y prevención.
